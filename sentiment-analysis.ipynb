```python
import numpy as np

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_score = {label: [] for label in emotion_labels}

def calculate_max_emotion(predictions):
    per_description_emotions = {label: [] for label in emotion_labels}

    for prediction in predictions:
         sorted_prediciton = sorted(prediction, key=lambda x:x["label"])
         for index, label in enumerate(emotion_labels):
              per_description_emotions[label].append(sorted_prediciton[index]["score"])

         return {label: np.max(scores) for label, scores in per_description_emotions.items()}
```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".")
    predictions = classifier(sentences)
    max_score = calculate_max_emotion(predictions)

    for label in emotion_labels:
        emotion_score[label].append(max_score[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_score = {label: [] for label in emotion_labels}

for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".")
    predictions = classifier(sentences)
    max_score = calculate_max_emotion(predictions)

    for label in emotion_labels:
        emotion_score[label].append(max_score[label])
```

```python
emotions_df = pd.DataFrame(emotion_score)
emotions_df["isbn13"] = isbn
```

```python
books = pd.merge(books, emotions_df, on="isbn13")
```

```python
books.to_csv("books_with_emotions.csv", index=False)
```
