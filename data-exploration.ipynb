```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# More the white more the missing values
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2026 - books["published_year"]
```

```python
# want to know the correlation of the missing_descripiton columns with the other columns as we are not going to use the entries with missing descriptions and whether the only books
# with less or more number of pages have their des missing could lead to having bias towards them

columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# spearman because we are dealing with binaries(non continuous values)
correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()

# We found no strong correlation
```

```python
missing_data = books[(books["age_of_book"].isna()) |
                  (books["num_pages"].isna()) |
                  (books["missing_description"].isna()) |
                  (books["average_rating"].isna())]
```

```python
books_missing = books[~(books["age_of_book"].isna()) &
                    ~(books["num_pages"].isna()) &
                    ~(books["missing_description"].isna()) &
                    ~(books["average_rating"].isna())]
```

```python
books_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
plt.figure(figsize=(12, 30))
books_missing["categories"].value_counts().sort_values(ascending=False).head(50).plot(kind="barh")
plt.xlabel("Total Count")
plt.ylabel("Category")
plt.title("Top 50 Categories")
plt.gca().invert_yaxis() # orders the columns in descending manner wrto their counts
plt.tight_layout()
plt.show()
```

```python
# Some of the descriptions are completly useless and does not tell anything about the book 
# have to sort them out and we can do it by checking how long they are
```

```python
books_missing["words_in_description"] = books_missing["description"].str.split().str.len()
```

```python
books_missing
```

```python
plt.figure(figsize=(9, 6))
sns.histplot(
    books_missing["words_in_description"],
    bins=30,
    color="blue",
    edgecolor="black",
    kde=True,
)

plt.show()
```

```python
# So the very short descriptions are concerning on whether they actually tells the descripition of the book or is just gibbersih

# For this we gonna look into various bins like 1-4, 5-14 etc
```

```python

books_missing.loc[books_missing["words_in_description"].between(1, 4), "description"]

# these are garbage descriptions are needs to be removed
```

```python
books_missing.loc[books_missing["words_in_description"].between(5, 14), "description"]
```

```python
# this range gives satisfactory information about the book 
books_missing.loc[books_missing["words_in_description"].between(15, 24), "description"]
```

```python
# Remove the books with description less than 25 words

book_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words.shape
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(),  book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].fillna("").astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
